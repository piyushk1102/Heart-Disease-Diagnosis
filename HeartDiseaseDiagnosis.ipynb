```python
target=data.groupby('target').size()
target
```

```python
def heart_disease(row):
    if row==0:
        return 'Absent'
    elif row==1:
        return 'Present'
```

```python
data['Heart_Disease']=data['target'].apply(heart_disease)
```

```python
def gender(row):
    if row==1:
        return 'Male'
    elif row==0:
        return 'Female'
```

```python
data['Gender']=data['sex'].apply(gender)
```

```python
#Pie Chart

import pandas as pd
import matplotlib.pyplot as plt

counts = data['Heart_Disease'].value_counts()

plt.figure(figsize=(20,20))
plt.pie(sizes, labels=['Present','Absent'], autopct='%0.0f%%', colors=['#66b3ff','#ff9999'], startangle=140)
plt.title('Heart Disease Population %', fontsize=20)
plt.show()
```

```python
#Countplot

import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))
sns.countplot(x='age', data=data)
plt.title('Population Age', fontsize=20)
plt.xlabel('Age', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
#Statistical Analysis

Min_Age=data['age'].min()
Max_Age=data['age'].max()
Mean_Age=data['age'].mean()
print("Minimum Age =",Min_Age)
print("Maximum Age =",Max_Age)
print("Mean Age =",Mean_Age)
```

```python
#Categorical Analysis

Young_Ages=data[(data['age']>=29) & (data['age']<40)]
Middle_Ages=data[(data['age']>=40) & (data['age']<55)]
Elderly_Ages=data[(data['age']>55)]
print('Young Ages =',len(Young_Ages))
print('Middle Ages =',len(Middle_Ages))
print('Elderly Ages =',len(Elderly_Ages))
```

```python
#Bar Plot Creation of Age Category

import matplotlib.pyplot as pyplot
import seaborn as sns

plt.figure(figsize=(10,7))
sns.barplot(x=['Young','Middle','Elderly'], y=[len(Young_Ages), len(Middle_Ages), len(Elderly_Ages)], palette='YlOrBr')
plt.title('Age Category', fontsize=20)
plt.xlabel('Age Range', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.show()
```

```python
def age_range(row):
    if row>=29 and row<40:
        return 'Young Age'
    elif row>=40 and row<55:
        return 'Middle Age'
    elif row>=55:
        return 'Elder Age'
```

```python
data['Age_Range']=data['age'].apply(age_range)
```

```python
data
```

```python
#Swarm Plot for Gender Based Age Category

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,20))
sns.swarmplot(x='Age_Range', y='age', hue='Gender', data=data, order=['Young Age','Middle Age','Elder Age'], palette='Oranges_r')
plt.title('Gender Based Age Category', fontsize=25)
plt.xlabel('Age Category', fontsize=20)
plt.ylabel('Age', fontsize=20)
plt.show()
```

```python
#Strip Plot for Gender Based Age Category

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,10))
sns.stripplot(x='Age_Range', y='age', hue='Gender', data=data, order=['Young Age','Middle Age','Elder Age'], palette='Oranges_r', dodge=True)
plt.title('Gender Based Age Category', fontsize=20)
plt.xlabel('Age Category', fontsize=15)
plt.ylabel('Age', fontsize=15)
plt.legend(title='Sex', loc='best')
plt.show()
```

```python
#Count Plot for Heart Disease Based On Age Category

plt.figure(figsize=(10,10))
hue_order=['Young Age', 'Middle Age', 'Elder Age']
sns.countplot(x='Heart_Disease', hue='Age_Range', data=data, order=['Present','Absent'], hue_order=hue_order, palette='Pastel1')
plt.title('Heart Disease Based On Age Category', fontsize=20)
plt.xlabel('Heart Disease', fontsize=15)
plt.ylabel('Counts', fontsize=15)
plt.show()
```
